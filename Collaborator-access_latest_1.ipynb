```python
import numpy as np
import pandas as pd
from selenium import webdriver
import random
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.common.by import By
from selenium.common.exceptions import WebDriverException
import time
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.maximize_window()
#Get url
driver.get('https://partners.shopify.com/637025/stores?status=collaborator_access&sort_order=created_at_desc&page=138')
time.sleep(5)
#Thay gmail
try:
    driver.find_element(By.XPATH,'//*[@id="account_email"]').send_keys('hangntt@bsscommerce.com')
    time.sleep(5)
    driver.find_element(By.XPATH,'//*[@id="body-content"]/div[2]/div/div[2]/div/div/div[2]/div/form/button').click()
    time.sleep(5)
    #send passwords
    driver.find_element(By.XPATH,'//*[@id="account_password"]').click()
    time.sleep(20)
    #click to login
    driver.find_element(By.XPATH,'//*[@id="login_form"]/div[2]/ul/button').click()
    time.sleep(10)
except WebDriverException:
    print("Login already")
```

```python
element_list = []
counter = 138
is_break = False
driver.get(f"https://partners.shopify.com/637025/stores?status=collaborator_access&sort_order=created_at_desc&page=138")
while True:
    if is_break:
        break
    store_nums = 0
    while True:
        try:
            store_nums = int(driver.find_element(By.XPATH,'//*[@id="AppFrameMain"]/div/div[1]/div[2]/div/div[2]/div[1]/div[2]/div[2]/div/div[2]/div/div/div[1]'
                           ).text[8:-7])
            break
        except NoSuchElementException:
            continue
    for i in range(1,store_nums+1):
        while True:
            try:
                name = driver.find_element("xpath",f'/html/body/div/div[2]/main/div[2]/div/div[1]/div/div[1]/main/div/div[1]/div[2]/div/div[2]/div[1]/div[2]/ul/li[{i}]/div/div/div/div/div/div[1]/p/span').text
                shopify_plan = driver.find_element("xpath",f'/html/body/div/div[2]/main/div[2]/div/div[1]/div/div[1]/main/div/div[1]/div[2]/div/div[2]/div[1]/div[2]/ul/li[{i}]/div/div/div/div/div/div[2]/div[1]/p').text
                link = driver.find_element("xpath",f'/html/body/div/div[2]/main/div[2]/div/div[1]/div/div[1]/main/div/div[1]/div[2]/div/div[2]/div[1]/div[2]/ul/li[{i}]/div/div/a').get_attribute('href')
                driver.get(link)
                time.sleep(3)
                try:
                    Store_domain = driver.find_element(By.XPATH,'//*[@id="AppFrameMain"]/div[2]/div/div[2]/div/section[1]/div[1]/p/a').get_attribute('href')
                except NoSuchElementException:
                    pass
                driver.back()
                time.sleep(3)
                element_item = {
                    'name':name,
                    'shopify_plan':shopify_plan,
                    'Store_domain':Store_domain
                    }
                element_list.append(element_item)
                break
            except NoSuchElementException:
                continue
    # check next button enabled or not ?
    next_pagination_cmt = 0
    while True:
        try:
            next_pagination_cmt = driver.find_element(By.XPATH,'//*[@id="AppFrameMain"]/div/div[1]/div[2]/div/div[2]/div[1]/div[3]/nav/button[2]')
            break
        except NoSuchElementException as e:
            print(str(e))
            continue
    print(next_pagination_cmt)
    if next_pagination_cmt.get_attribute("disabled") == "true":
        is_break = True
        break
    else:
        next_pagination_cmt.click()
        counter += 1
        time.sleep(2)


```
