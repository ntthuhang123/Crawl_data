```python
import numpy as np
import pandas as pd
from selenium import webdriver
from time import sleep
import random
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.common.by import By
# Declare brower
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://apps.shopify.com/b2b-solution-custom-pricing/reviews')
```

```python
reviewer_name,location,date,review_content,rating = [], [],[],[],[]
count = 1
while True:
    try:
        driver.get('https://apps.shopify.com/b2b-solution-custom-pricing/reviews?page='+str(count))
    
        elems_name = driver.find_elements(By.CSS_SELECTOR, ".review-listing-header")
        if len(elems_name) == 0:
            break
        reviewer_name = reviewer_name + [elem.text for elem in elems_name]  
        
        elems_location = driver.find_elements(By.CSS_SELECTOR, ".review-merchant-characteristic__item")
        location = location + [elem.text for elem in elems_location] 
        
        elems_date = driver.find_elements(By.CSS_SELECTOR, ".review-metadata__item-label")
        date = date + [elem.text for elem in elems_date] 
        
        elems_content = driver.find_elements(By.CSS_SELECTOR,".review-content")
        review_content = review_content + [elem.text for elem in elems_content] 
        
        elems_rating = driver.find_elements(By.CSS_SELECTOR, ".review-metadata__item")
        rating = rating + [elem.text for elem in elems_rating] 
 
        print(f"clicked on button next page {count} ")
        sleep(random.randint(1,3))
        count += 1
    except ElementNotInteractableException:
        print("Element Not Interactable Exception!")
        break
 
```

```python
location_after_clean = []
for i in location:
    if i.startswith("Location"):
        location_raw = i.split()[1:]
        location_right = ""
        for index,j in enumerate(location_raw):
            if index > 0:
                location_right = location_right + " " + j
            else:
                location_right += j
        location_after_clean.append(location_right)    
           
        
```

```python
df_1 = pd.DataFrame(list(zip(reviewer_name,location_after_clean,date,review_content,rating)),columns= ['reviewer_name','location','date','review_content','rating'])
df_1
```

```python
time_idx = []
time_using = []
h = 1
count = 1
is_break = False
while True:

    driver.get('https://apps.shopify.com/b2b-solution-custom-pricing/reviews?page='+str(count))
    for j in range(2,12):
        try:
            time_elem = driver.find_element("xpath","/html/body/div[7]/main/div/section/div[2]/div[2]/div[{}]/div[1]/div[3]/div[2]".format(j))
                                                     
            time_using.append(time_elem.text)
            
            time_idx.append(h) 
            h += 1   
        except NoSuchElementException: 
            is_break = True
            print(" No such Element Exception" + str(j))
            break  
    if is_break:
        break
    print(f"clicked on button next page {count} ") 
    count = count + 1        
       
```

```python
time_spent = []
for i in time_using:
    time_raw = i.split("app")[1]
    time_spent.append(time_raw.strip())
   
```

```python
time_spent_clean = []
for i in time_spent:
    if i.find("About") != -1:
        a = i.replace("About","")
        time_spent_clean.append(a)
    elif i.find("Over") != -1:
        b = i.replace("Over","")
        time_spent_clean.append(b)
    else:
        time_spent_clean.append(i)
print(time_spent_clean)    
len(time_spent_clean)  
```

```python
date_clean = []
for i in date:
    if i.find("Edited") != -1:
        a = i.replace("Edited","")
        date_clean.append(a.strip())
    else:
        date_clean.append(i)
print(date_clean)        
```

```python
#Convert date_clean to format mm/dd/yyyy
from datetime import datetime
date_convert = []

def mdy_to_ymd(d):
    return datetime.strptime(d,'%B %d, %Y').strftime('%m/%d/%Y')
for i in date_clean:
    date_ = mdy_to_ymd(i)
    date_convert.append(date_)

print(date_convert)
```

```python
days_using_app = []
for i in time_spent_clean:
    if i.find("minute") != -1:
        a = int(i.split()[0])/1440
        days_using_app.append(a)   
    elif i.find("month") != -1:
        b = int(i.split()[0])*30
        days_using_app.append(b)
    elif i.find("year") != -1:
        c = int(i.split()[0])*365
        days_using_app.append(c)
    elif i.find("hour") != -1:
        d = int(i.split()[0])/24
        days_using_app.append(d)
    else:
        e = int(i.split()[0])
        days_using_app.append(e)
print(days_using_app)        
len(days_using_app) 
```
