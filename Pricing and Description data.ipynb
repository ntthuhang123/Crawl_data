```python
import gspread
import numpy as np
import pandas as pd
from selenium import webdriver
import random
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.common.by import By
import time
```

```python
sa = gspread.service_account(filename="shopify-market-research-3542cb4db6df.json")
sh = sa.open("kw_reaseach")
wks = sh.worksheet("competitor")
```

```python
link_competitor = wks.get('A2:A')
```

```python
driver = webdriver.Chrome('./chromedriver.exe')
driver.maximize_window()
element_list = []
for i in link_competitor:
    for j in i:
        driver.get(j)
        time.sleep(2)
        try: 
            app_name = driver.find_element(By.XPATH,'//*[@id="adp-hero"]/div/div/div[1]/div/div[1]/div[1]/h1').text
            owner = driver.find_element(By.XPATH,'//*[@id="adp-hero"]/div/div/div[1]/div/div[1]/div[2]/div[2]/div[3]/div').text
            description = driver.find_element(By.XPATH,'//*[@id="adp-details-section"]/div/div/div[2]').text
            pricing = driver.find_element(By.XPATH,'//*[@id="adp-pricing"]/div/div/div/div[2]').text
        except NoSuchElementException:
            print("No element")
    
        element_item = {
            'App_name':app_name,
            'Link' : j,
            'owner':owner,
            'full_description' : description,
            'pricing_plan': pricing
        }
        
        element_list.append(element_item)

    
```

```python
df_1 = pd.DataFrame(element_list)
df_1
```

```python
from datetime import date 
# Returns the current local date
today = date.today()
print("Today date is: ", today)
```

```python
df_1.insert(0,'Date',today)
```

```python
bss_app = wks.get('B2:B')
bss_list = []
for i in bss_app:
    for j in i:
        bss_list.append(j)
df_1.insert(1,'BSS_app',bss_list)
df_1     
```

```python
#Import data into file pricing
pricing_df = df_1.iloc[:,[0,1,2,3,4,6]]
import gspread
from gspread_dataframe import set_with_dataframe
from google.oauth2.service_account import Credentials
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
import pygsheets

client = pygsheets.authorize(service_account_file="shopify-market-research-3542cb4db6df.json")
sh=client.open('Linh - Full Shopify App Market Research')
wks = sh.worksheet_by_title('Pricing Data') # open the existing file
cells = wks.get_all_values(include_tailing_empty_rows=False, include_tailing_empty=False, returnas='matrix')
last_row_1 = len(cells)+1
wks.set_dataframe(pricing_df,(last_row_1,0),copy_head = False,extend = True )
```
