```python
import numpy as np
import pandas as pd
from selenium import webdriver
import random
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
from selenium.webdriver.common.by import By
# Declare brower
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get('https://apps.shopify.com/b2b-solution-custom-pricing/reviews')
```

```python
count = 1
review_list = []
while True:
    try:
        driver.get('https://apps.shopify.com/b2b-solution-custom-pricing/reviews?page='+str(count))
        reviews = driver.find_elements(By.CLASS_NAME,"review-listing ")
        if len(reviews) == 0:
            break
        for review in reviews:
            name = review.find_element(By.CSS_SELECTOR, ".review-listing-header").text   
            location = review.find_element(By.CSS_SELECTOR, ".review-merchant-characteristic__item span").text
            time_spent = review.find_element(By.CSS_SELECTOR,"div.review-merchant-characteristic__item:nth-of-type(2) span").text
            date = review.find_element(By.CSS_SELECTOR, ".review-metadata__item-label").text
            content = review.find_element(By.CSS_SELECTOR,".review-content").text
            rating = review.find_element(By.CSS_SELECTOR, ".review-metadata__item").text

            Review_item = {
            'name':name,
            'location':location,
            'time_spent':time_spent,
            'date':date,
            'content':content,
            'rating':rating
            }        
            review_list.append(Review_item)
        count += 1
        print(f"clicked on button next page {count} ")
    except NoSuchElementException:
        print(" No such Element Exception")
        break
    

 
```

```python
len(review_list)
```

```python
df_1 = pd.DataFrame(review_list)
df_1
```

- Clean data

```python
# clean rating column
df_1['rating_clean'] = df_1['rating'].apply(lambda x: pd.Series(str(x).split()))[0] 
del df_1['rating']
#Rename rating_clean column
df_1.rename(columns = {'rating_clean': 'rating'},inplace = True )
# clean time_spent column
for i in df_1.index:
    if df_1.loc[i,'time_spent'].find('About') != -1:
        df_1.loc[i,'time_spent'] = df_1.loc[i,'time_spent'].replace('About',"").strip()
    elif df_1.loc[i,'time_spent'].find('Over') != -1:
        df_1.loc[i,'time_spent'] = df_1.loc[i,'time_spent'].replace('Over',"").strip()
# clean date column
for i in df_1.index:
    if df_1.loc[i,'date'].find('Edited') != -1:
        df_1.loc[i,'date'] = df_1.loc[i,'date'].replace('Edited',"").strip()
```

```python
df_1
```

```python
#Caculate days of using app
days_using = []
for i in df_1.time_spent:
    if i.find("minute") != -1:
        a = int(i.split()[0])/1440
        days_using.append(a)   
    elif i.find("month") != -1:
        b = int(i.split()[0])*30
        days_using.append(b)
    elif i.find("year") != -1:
        c = int(i.split()[0])*365
        days_using.append(c)
    elif i.find("hour") != -1:
        d = int(i.split()[0])/24
        days_using.append(d)
    else:
        e = int(i.split()[0])
        days_using.append(e)
df_1['days_using'] = days_using
```
